```python
# Bumrah dataset
bumrah_df = df[df.bowler == 'JJ Bumrah']
```

```python
bumrah_df.head()
```

```python
# Runs Scored by every batsman against Bumrah
# Assigning a name to this particular dataset
# 'groupby' converts the dataframe into a series, therefore reset index converts the data from being present under an index to being present under a column
bdf1 = pd.DataFrame(bumrah_df.groupby('striker')['runs_off_bat'].sum()).reset_index()
```

```python
# Result of conversion - data is presented within a table under given column names
bdf1.head()
```

```python
# Balls Faced by each batsman against Bumrah
# Assigning a name to this dataset
# Same conversion as above
bdf2 = pd.DataFrame(bumrah_df.groupby('striker')['ball'].count()).reset_index()
```

```python
bdf2.head()
```

```python
# merging the striker column with 'runs' and 'balls' in the same table together
# Assigning a new name to this combined dataset
bdf3 = bdf1.merge(bdf2, on = 'striker', how = 'left')
```

```python
# Adding a new field 'strike rate' in the combined dataset
bdf3['strike rate'] = 100*(bdf3['runs_off_bat']/bdf3['ball'])
```

```python
bdf3.head()
```

```python
# adding a minimum criteria of 'balls faced' >=30 to get rid of outliers and have a more reliable data
bdf3 = bdf3[bdf3.ball >= 30]
```

```python
bdf3.head()
```

```python
# ABD Dataset
abd_df = df[df.striker == 'AB de Villiers']
```

```python
abd_df.head()
```

```python
# Runs scored by ABD against each bowler
# Assigning a name to this dataset
abdf1 = pd.DataFrame(abd_df.groupby('bowler')['runs_off_bat'].sum()).reset_index()
```

```python
abdf1.head()
```

```python
# Balls faced by ABD against each bowler
abdf2 = pd.DataFrame(abd_df.groupby('bowler')['ball'].count()).reset_index()
```

```python
abdf2.head()
```

```python
# merging the striker column with 'runs' and 'balls' in the same table together
# Assigning a new name to this combined dataset
abdf3 = abdf1.merge(abdf2, on = 'bowler', how = 'left')
```

```python
abdf3.head()
```

```python
# Adding a new field 'strike rate' in the combined dataset
abdf3['strike rate'] = 100*(abdf3['runs_off_bat']/abdf3['ball'])
```

```python
abdf3.head()
```

```python
# Minimum criteria of 'Balls Faced' >=30
abdf3 = abdf3[abdf3.ball >= 30]
```

```python
abdf3.head()
```

```python
# bdf3, abdf3
```

```python
# To reorder in the index nos. in the new filtered table
bdf3.reset_index(inplace = True , drop = True)
abdf3.reset_index(inplace = True , drop = True)
```

```python
plt.scatter(bdf3['strike rate'], bdf3['runs_off_bat'])
for i in range (len(bdf3)):
    # plt.text(x,y,text)
    if bdf3['striker'][i] == 'AB de Villiers':
        plt.scatter(bdf3['strike rate'][i], bdf3['runs_off_bat'][i],c = 'red')
        plt.text(bdf3['strike rate'][i] -32, bdf3['runs_off_bat'][i], bdf3['striker'][i])
    else:
       plt.text(bdf3['strike rate'][i]+5, bdf3['runs_off_bat'][i], bdf3['striker'][i])
plt.axvline(130, ls = '--',color = 'yellow')
plt.axhline(70,  ls = '--',color = 'yellow')
plt.xlabel('Strike Rate')
plt.ylabel('Runs Scored')
plt.title('Bumrah v/s All Batters in IPL (Min. 30 Balls Faced)', fontsize = 16)
plt.figure(figsize = (50,30))
plt.show()
```
